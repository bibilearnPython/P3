```python
##### 2 pages ######

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import random
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

### permet de faire des pauses pour charger le contenu de la page ##
def random_delay():
    time.sleep(random.uniform(2, 5))


### initialisation du driver chrome  ###
path = '/Users/jaroldmartinez/Documents/CODE/chromedriver'
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)


data_frames = []

for i in links_page:                                               # boucle pour traiter toutes les pages 
    random_delay()
    driver.get(i)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    job_page = soup.findAll(class_='css-5lfssm eu4oa1w0')          # cette classe correspond aux bloc contenant toutes les annonces
    list_of_links = []
    for job in job_page:                                           # je vais chercher le lien de chaque annonce que je stocke dans la liste 'list_of_links'
        j = job.find_all(class_='jcs-JobTitle css-jspxzf eu4oa1w0')
        for a in j:
            list_of_links.append(a['href'])

    for url in list_of_links:                                                                        
        full_url = 'https://fr.indeed.com/'+url                   
        driver.get(full_url)                                     
        random_delay()
        soup_1 = BeautifulSoup(driver.page_source, 'html.parser')  

        job_title = soup_1.find(class_='jobsearch-JobInfoHeader-title css-1b4cr5z e1tiznh50')        # le titre de l'annonce 
        job_title_text = job_title.text if job_title is not None else np.nan

        company_name = soup_1.find(class_='css-1saizt3 e1wnkr790')                                   # nom de l'entreprise 
        company_name_text = company_name.text if company_name is not None else np.nan

        localisation = soup_1.find(class_='css-waniwe eu4oa1w0')                                     # lieu du poste 
        localisation_text = localisation.text if localisation is not None else np.nan

        informations = soup_1.findAll(class_='js-match-insights-provider-36vfsm eu4oa1w0')           # les informations sous forme de catégories: pour chaque annonce il y a parfois des rubriques (comme compétences, types de poste ..) 
                                                                                                     # qui contiennent une ou plusieurs infos   
        section_elements = {}
        for info in informations:                                                                    # 'informations' correspond au bloc contenant toute les rubriques
            titles = info.findAll(class_='js-match-insights-provider-e6s05i eu4oa1w0')               # cette boucle va aller chercher les lignes où nous trouvons les info pour chaque rubrique

            for f in titles:                                                                         # pour chaque rubrique ..
                random_delay()
                h3_element = f.findAll('h3', class_='js-match-insights-provider-11n8e9a e1tiznh50')  # h3 correspond au titre de la rubrique
                res = f.findAll(class_='js-match-insights-provider-tvvxwd ecydgvn1')                 # res correspond au différents résultat de chaque rubrique

                for h3 in h3_element:                                                                # pour chaque titre..
                    categorie = h3.text                                                              # dans une variable je prend .text pour n'avoir que le titre

                    for r in res:                                                                    # pour chaque resultat ..
                        section_elements.setdefault(categorie, []).append(r.text)                    # j'ajoute dans mon dictionnaire le titre en clé et en valeurs une liste contenant le contenu de chaque rubrique

        description = soup_1.find(class_='jobsearch-JobComponent-description css-16y4thd eu4oa1w0')  # la description du poste 
        description_text = description.text if description is not None else np.nan



        
        dicto = {'poste': job_title_text, 'entreprise': company_name_text, 'lieu du poste': localisation_text, 'description': description_text,'lien du poste':full_url}
        full = dicto | section_elements

        data_1 = pd.DataFrame(dicto, index=[0])
        series = pd.Series(section_elements)
        series = series[series != '']
        data_2 = series.to_frame().T
        data = pd.concat([data_1, data_2], axis=1)
        data_frames.append(data)

        print(f"Scraped data for job: {job_title_text}")

full_df = pd.concat(data_frames, ignore_index=True)
print(full_df)
```
